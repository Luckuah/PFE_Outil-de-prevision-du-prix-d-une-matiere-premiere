```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

def download_data(ticker: str, start: str, end: str) -> pd.DataFrame:
    """Télécharge les données depuis Yahoo Finance."""
    data = yf.download(ticker, start=start, end=end)
    return data

def add_indicators(data: pd.DataFrame, ma_windows=[30]) -> pd.DataFrame:
    """Ajoute moyenne mobile(s) et rendement journalier sur le Close."""
    for window in ma_windows:
        data[f"MA{window}"] = data["Close"].rolling(window).mean()
    data["Returns"] = data["Close"].pct_change()
    return data

def normalize_data(data: pd.DataFrame) -> pd.DataFrame:
    """Normalise toutes les colonnes numériques entre 0 et 1."""
    numeric_cols = data.select_dtypes(include="number").columns
    scaler = MinMaxScaler()
    data[numeric_cols] = scaler.fit_transform(data[numeric_cols])
    return data

def save_csv(data: pd.DataFrame, filename: str):
    """Sauvegarde les données dans un CSV."""
    data.to_csv(filename)
    print(f"✅ Données sauvegardées dans {filename}")

def plot_data(data: pd.DataFrame, columns=["Close", "MA30"]):
    """Visualise les colonnes sélectionnées."""
    plt.figure(figsize=(12,6))
    for col in columns:
        if col in data.columns:
            plt.plot(data[col], label=col)
    plt.title("Évolution des prix normalisés")
    plt.xlabel("Date")
    plt.ylabel("Valeur normalisée")
    plt.legend()
    plt.show()

# -----------------------------
# Exemple d'utilisation
# -----------------------------
if __name__ == "__main__":
    ticker = "CL=F"
    start_date = "1960-01-01"
    end_date = "2025-01-01"

    df = download_data(ticker, start_date, end_date)
    df = add_indicators(df, ma_windows=[7, 30, 90])
    df = normalize_data(df)
    save_csv(df, "wti_oil_normalized.csv")
    plot_data(df, columns=["Close", "MA7", "MA30", "MA90"])
```

```python
def future_signal(data: pd.DataFrame, open_col="Open", close_col="Close") -> pd.DataFrame:
    """
    Ajoute des colonnes signal pour 7 et 14 jours basées sur la variation future par rapport à l'Open du jour.
    """
    # Différence future
    data["diff1"] = (data[close_col].shift(-1) - data[open_col]) / data[open_col]
    data["diff7"] = (data[close_col].shift(-14) - data[open_col]) / data[open_col]

    # Fonction pour appliquer la règle de signal
    def signal(x):
        if x > 0.01:
            return 1
        elif x < -0.01:
            return -1
        else:
            return 0

    # Appliquer la fonction signal
    data["signal1"] = data["diff1"].apply(signal)
    data["signal7"] = data["diff7"].apply(signal)

    # Supprimer les colonnes intermédiaires si tu veux
    # data.drop(columns=["diff7", "diff14"], inplace=True)

    return data

# -------------------------
# Exemple d'utilisation
# -------------------------
if __name__ == "__main__":
    import yfinance as yf

    df = yf.download("CL=F", start="2020-01-01", end="2025-10-01")
    df = future_signal(df)
    df_period = df.loc["2025-08-12":"2025-09-12"]

    # Afficher
    print(df_period[["Open", "Close", "signal1", "signal7"]])


```

```python
target_col = "signal1"  # ta colonne signal 1
feature_cols = [col for col in df.columns if col != target_col]

# Convertir en numpy
features = df[feature_cols].values
target = df[target_col].values

# Paramètres
sequence_length = 30  # nombre de jours pour une séquence
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
class TimeSeriesDataset(Dataset):
    def __init__(self, X, y, seq_len=30):
        self.X = X
        self.y = y
        self.seq_len = seq_len

    def __len__(self):
        return len(self.X) - self.seq_len

    def __getitem__(self, idx):
        seq_x = self.X[idx:idx+self.seq_len]
        seq_y = self.y[idx+self.seq_len]
        return torch.tensor(seq_x, dtype=torch.float32), torch.tensor(seq_y, dtype=torch.float32)

dataset = TimeSeriesDataset(features, target, sequence_length)
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
```
