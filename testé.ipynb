```python
import numpy as np

# Rendements log quotidiens
df["Returns"] = np.log(df["Close"] / df["Close"].shift(1))

# Volatilité hebdomadaire (écart-type des rendements sur 5 jours ouvrés)
weekly_vol = df["Returns"].rolling(5).std() * np.sqrt(5)

df["WeeklyVol"] = weekly_vol
print(df[["Close", "Returns", "WeeklyVol"]].head(15))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
N = 90        # nombre de jours simulés (au lieu de semaines)
dt = 1        # 1 pas = 1 jour
S0 = df["Close"].iloc[-1]  # dernier prix comme point de départ

# Dérive et volatilité journalière
# Si tes Returns sont journaliers, on n’a pas besoin de multiplier par 5
mu = df["Returns"].mean()      # dérive journalière moyenne
sigma = df["Returns"].std()    # volatilité journalière

n_paths = 300   # nombre de trajectoires

# Simulation
np.random.seed(42)
Z = np.random.normal(0, 1, size=(n_paths, N))
increments = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
log_paths = np.cumsum(increments, axis=1)
paths = S0 * np.exp(log_paths)

# Calcul de la trajectoire moyenne
mean_path = paths.mean(axis=0)

# Affichage
plt.figure(figsize=(12,6))
for i in range(n_paths):
    plt.plot(range(N), paths[i], color='lightgray', lw=1)
plt.plot(range(N), mean_path, color='red', lw=2, label='Trajectoire moyenne')
plt.plot(range(N), df2["Close"].values[:N], color='blue', lw=2, label='Prix réels')
plt.title("30 trajectoires simulées journalières avec trajectoire moyenne")
plt.xlabel("Jours")
plt.ylabel("Prix simulé")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# paths : shape (n_paths, N)
# N : nombre de pas simulés

# 1️⃣ Espérance (moyenne de toutes les trajectoires)
esperance = paths.mean(axis=0)

# 2️⃣ Moyenne des 5% trajectoires les plus basses
low_idx = int(np.ceil(0.05 * n_paths))  # nombre de trajectoires à considérer
high_idx = int(np.ceil(0.05 * n_paths))

# Trier les trajectoires pour chaque pas
sorted_paths = np.sort(paths, axis=0)  # tri par colonne (chaque pas)
low_5 = sorted_paths[:low_idx, :].mean(axis=0)
high_5 = sorted_paths[-high_idx:, :].mean(axis=0)

# 3️⃣ Affichage
plt.figure(figsize=(12,6))

# Trajectoires individuelles
for i in range(n_paths):
    plt.plot(range(N), paths[i], color='lightgray', lw=1)

# Espérance
plt.plot(range(N), esperance, color='red', lw=2, label='Espérance')

# 5% low et 5% high
plt.plot(range(N), low_5, color='blue', lw=2, linestyle='--', label='Moyenne 5% low')
plt.plot(range(N), high_5, color='green', lw=2, linestyle='--', label='Moyenne 5% high')
plt.plot(range(N), df2["Close"].values[:N], color='purple', lw=2, label='Prix réels')

plt.title("Simulation avec espérance et 5% low/high")
plt.xlabel("Pas")
plt.ylabel("Prix simulé")
plt.legend()
plt.grid(True)
plt.show()
```
